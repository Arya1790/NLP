```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
import re
import nltk
from nltk.stem import WordNetLemmatizer
import pandas as pd
```

```python
nltk.download('wordnet')
nltk.download('punkt')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
messages = pd.read_csv("/content/drive/MyDrive/Dataset/SMSSpamCollection.txt",
                       sep='\t', names=["label", "message"])
messages.head()
```

```python
corpus = []
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review]
    review=' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus[0:5]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
  sent_tokens = sent_tokenize(sent)
  for sent_token in sent_tokens:
    words.append(simple_preprocess(sent_token))

print(words[0])
print(words[1])
print(words[2])
```

```python
# word2vec from scratch
# vector_size=100 dimensions
model = gensim.models.Word2Vec(words,vector_size=100,epochs=50)
```

```python
# vocabulary
model.wv.key_to_index
```

```python
# vocabulary size
model.corpus_count
```

```python
model.epochs
```

## Avg Word2Vec

the entire sentence will have dimension 100

```python
from tqdm import tqdm
import numpy as np
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
#apply for the entire sentences
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
##independent Features
print(X[0].shape)
X[0]
```

```python
## Dependent Features
## Output Features
y = messages[list(map(lambda x: len(x)>0 ,corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values
```

```python
## this is the final independent features
df=pd.DataFrame()
df_list = []
for i in range(0,len(X)):
    df_list.append(pd.DataFrame(X[i].reshape(1,-1)))

df = pd.concat(df_list, ignore_index=True)
```

```python
df['Output']=y
df.dropna(inplace=True)
df.head()
```

```python
## Independent Feature
X=df.iloc[:,:-1]
## Dependent Feature
y=df['Output']
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)

y_pred=classifier.predict(X_test)
```

```python
y_pred=classifier.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
