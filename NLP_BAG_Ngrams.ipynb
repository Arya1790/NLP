```python
messages = pd.read_csv("/content/drive/MyDrive/Dataset/SMSSpamCollection.txt",
                       sep='\t', names=["label", "message"])
messages.head()
```

```python
# data cleaning
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[0]
```

## BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
# select top 2500 words, binary bag of words
cv = CountVectorizer(max_features=2500, binary=True)
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
# converting words into vectors
X
```

```python
# words and index's
cv.vocabulary_
```

## N-grams

```python
# select top 500 words, binary bag of words, n grams with uni gram and bi gram
cv = CountVectorizer(max_features=500, binary=True,ngram_range= (1,2))
X = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
# bi gram
cv = CountVectorizer(max_features=100, binary=True,ngram_range= (2,2))
X = cv.fit_transform(corpus).toarray()
```
